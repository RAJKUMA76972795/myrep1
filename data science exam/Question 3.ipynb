```python
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/Shopping_Revenue.csv")
```

```python
data
```

```python
data.info()
```

```python
y=data.loc[:,"revenue"]
data.drop(columns="revenue",inplace=True)
y
```

```python
X=data.iloc[:,5:]
X
```

```python
data_city=pd.get_dummies(data["City"],drop_first=True)
data_city
```

```python
data_city_group=pd.get_dummies(data["City Group"],drop_first=True)
data_city_group
```

```python
data_Type=pd.get_dummies(data["Type"],drop_first=True)
data_Type
```

```python
X=pd.concat([X,data_city_group,data_Type],axis=1)
X
```

```python
X_train , X_test , y_train , y_test= train_test_split(X,y,test_size=0.2,random_state=51)
```

```python
regressor=RandomForestRegressor()
regressor.fit(X_train,y_train)
```
